```python
import requests 
from bs4 import BeautifulSoup 


def addUrls(url, listUrl): 
    #open with GET method 
    resp=requests.get(url)         
    if resp.status_code==200: 
        print("Successfully opened the web page") 
        listUrl.append(url)
    else:
        print("Did not work")
    soup=BeautifulSoup(resp.text,'html.parser')
    l=soup.find("a",{"class":"_2glka"})   #_2glka has to be adjusted based on website  
    if (l.get('href') == ""):
        print('Done adding')
    else:                
        addUrls(l.get('href'), listUrl) 


```

```python
def addSubSites(url, subSiteList): 
    #open with GET method 
    resp=requests.get(url)         
    if resp.status_code==200: 
        print("Successfully opened the web page") 
        #listUrl.append(url)
    else:
        print("Did not work")
    soup=BeautifulSoup(resp.text,'html.parser')
    l=soup.findAll("div",{"data-tid":"listing-info"})
    
    for listing in l:
        for link in listing:
            if(link.get('href')):
                subSiteList.append('https://www.apartmentguide.com'+link.get('href'))
                print('Following site added: {}'.format(link.get('href')))
        print('###')
```

```python
url = 'https://www.apartmentguide.com/apartments/Ohio/Dayton/'
listUrl = []
addUrls(url, listUrl)
```

```python
subSiteUrls = []
for url in listUrl:
    addSubSites(url, subSiteUrls)
```

```python
def gatherListingInfo(url):
    global globCount
    globCount = globCount + 1
    #open with GET method 
    resp=requests.get(url)         
    if resp.status_code==200: 
        print("Successfully opened the web page") 
        #listUrl.append(url)
    else:
        print("Did not work")
    soup=BeautifulSoup(resp.text,'html.parser')
    
    
    print(soup.find('h1',{'data-tid':'property-title'}).text)
    l=soup.findAll("div",{"data-tag_item":"floor_plans"})
    #for every sub layer of div tags loop;
    for i in l:
        print("###General information for apartment(s)###")
        for j in i.findAll('div',{"data-tid":"floorplan-row"}):
            for p in j:
                for prop in p:
                    mylist = []
                    for x in prop.findChildren('span', recursive = False):
                        mylist.append(x.text)
                        
                    if len(mylist) == 0:
                        print(prop.text)
                    else:
                        for item in mylist:
                            print(item)
                      
        print("\n###Done with general information###")
        count = 1
        for k in i.findAll('div',{'data-tag_item':"unit_title"}):
            print('\n#Unit {}:\n'.format(count))
            for g in k:
                mylist = []
                for yolo in g.findChildren('span', recursive=False):
                    mylist.append(yolo.text)
                
                if len(mylist) == 0:
                    print(g.text)
                else:
                    for item in mylist:
                        print(item)
            count = count + 1
        print('\n')
    print("\n###Done with Website number {}###\n".format(globCount))          
```

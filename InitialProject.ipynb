```python
import requests 
from bs4 import BeautifulSoup 


def addUrls(url, listUrl): 
    #open with GET method 
    resp=requests.get(url)         
    if resp.status_code==200: 
        print("Successfully opened the web page") 
        listUrl.append(url)
    else:
        print("Did not work")
    soup=BeautifulSoup(resp.text,'html.parser')
    l=soup.find("a",{"class":"_2glka"})   #_2glka has to be adjusted based on website  
    if (l.get('href') == ""):
        print('Done adding')
    else:                
        addUrls(l.get('href'), listUrl) 


```

```python
def addSubSites(url, subSiteList): 
    #open with GET method 
    resp=requests.get(url)         
    if resp.status_code==200: 
        print("Successfully opened the web page") 
        #listUrl.append(url)
    else:
        print("Did not work")
    soup=BeautifulSoup(resp.text,'html.parser')
    l=soup.findAll("div",{"data-tid":"listing-info"}) #_2glka has to be adjusted based on website  
    
    for listing in l:
        for link in listing:
            if(link.get('href')):
                subSiteList.append(link.get('href'))
                print('Following site added: {}'.format(link.get('href')))
        print('###')
```

```python
url = 'https://www.apartmentguide.com/apartments/Ohio/Dayton/'
listUrl = []
addUrls(url, listUrl)
```

```python
subSiteUrls = []
for url in listUrl:
    addSubSites(url, subSiteUrls)
```
